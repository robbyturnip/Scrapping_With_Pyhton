```python
import requests
import csv
from bs4 import BeautifulSoup
data=[] #inisialisasi array
kolom=[]#inisialisasi array
kolom.append(["Chapter","Starter Post","Time Post","Views","Replies"]) #nama kolom
location = "https://opforum.net/past-chapters/page-" #url situs , dimana seluruh data dibagi dalam sebuah halaman 

for page in range(1,13):#perulangan untuk mendapatkan angka dari 1-13 pada page
    minta= requests.get(location+str(page)) #memanggil url dengan menambahkan angka dibelakang url untuk mengatasi data pada halaman selanjutnya
    kirim=BeautifulSoup(minta.text,"html.parser")#parsing data html 
    produk=kirim.select("li.discussionListItem.visible")#memilih komponen html --> li dengan class discussionListItem visible
    for item in produk :#perulangan untuk mendapatkan data pada setiap halaman 
        chapter = item.select("a.PreviewTooltip")[0].text #memilih komponen html --> a dengan class PreviewTooltip
        poster = item.select("a.username")[0].text #memilih komponen html --> a dengan class username
        time=item.select("span.DateTime")[0].text #memilih komponen html --> span dengan class DateTime
        replie=item.select("dl.major > dd")[0].text  #memilih komponen html --> dl dengan class major dimana didalamnya terdapat komponen dd
        view=item.select("dl.minor > dd")[0].text #memilih komponen html --> dl dengan class minor dimana didalamnya terdapat komponen dd
        print(chapter," "+poster," "+time," "+replie," "+view,"\n") #cetak hasil pengambilan data melalui html
        data.append([chapter.strip(),poster,time,view,replie]) #menyimpan hasil tersebut ke dalam array data
    
        
with open("chapter_one_piece.csv","w")as filew: #membuat file csv
    writer=csv.writer(filew)
    writer.writerows(kolom)
    writer.writerows(data)
    
```
